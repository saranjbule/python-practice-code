<b>Operations</b>

- ENQUEUE (Inserting into a Queue)
- DEQUEUE (Deleting from a Queue)
- FRONT (Accessing the front element of a Queue)
- REAR (Accessing the rear element of a Queue)
- SIZE (Total elements in a Queue)
- isEmpty (Boolean value to define if Queue is empty)
<br /><br />
- ORDER of operation in Queue is FIFO (First In First Out)

<h3>Queue Implementation using Array/List

```python
class Queue:
    def __init__(self):
        self.__queue = []
        self.__front = -1
        self.__rear = -1
        self.__size = 0

    def enqueue(self, value):
        self.__queue.append(value)

        if self.__front == -1 and self.__rear == -1:
            self.__front = self.__rear = 0
        else:
            self.__rear += 1

        self.__size += 1

    def dequeue(self):
        if self.isEmpty():
            print('ERROR: Queue is empty')
            return
        # 1
        # firstElement = self.__queue.pop(0) # O(n) operation
        # 2
        firstElement = self.__queue[self.__front] # not deleting the element
        
        if self.__front == self.__rear:
            self.__front = self.__rear = -1
        else:
            self.__front += 1

        self.__size -= 1

        return firstElement

    def size(self):
        return self.__size
    
    def isEmpty(self):
        if self.__front == -1 and self.__rear == -1:
            return True
        
        return False
    
    def front(self):
        if self.isEmpty():
            print('ERROR: Queue is empty')
            return
        
        return self.__queue[self.__front]
    
    def rear(self):
        if self.isEmpty():
            print('ERROR: Queue is empty')
            return
        
        return self.__queue[self.__rear]
```

<h3>Queue Implemenatition using Linkedlist

```python
class Node:
    def __init__(self, val, next=None):
        self.val = val
        self.next = next

class Queue:
    def __init__(self):
        self.__head = None
        self.__tail = None
        self.__size = 0

    def enqueue(self, value):
        n = Node(value)

        if self.__head is None:
            self.__head = self.__tail = n
        else:
            self.__tail.next = n
            self.__tail = n

        self.__size += 1

    def dequeue(self):
        if self.isEmpty():
            print('ERROR: Queue is empty')
            return
        
        current_head = self.__head
        self.__head = self.__head.next
        self.__size -= 1

        return current_head.val

    def size(self):
        return self.__size

    def isEmpty(self):
        if self.__head is None or self.__size == 0:
            return True
        
        return False
    
    def front(self):
        if self.isEmpty():
            print('ERROR: Queue is empty')
            return
        
        return self.__head.val
    
    def rear(self):
        if self.isEmpty():
            print('ERROR: Queue is empty')
            return
        
        return self.__tail.val
```
